## Write a function is_strong(password) that checks if the password:

* Has at least 8 characters

* Contains numbers

* Contains uppercase letters

* Contains lowercase letters

### Return True or False.

```python
def strong(password):
    return (
        len(password) >= 8
        and any(c.isdigit() for c in password)
        and any(c.isupper() for c in password)
        and any(c.islower() for c in password)
    )
print(strong("anas1234"))
```

## Airline Seat Booking With Constraints

### Write a function book_seat(seats, seat_requested):

### Constraints:

    If seat_requested not in seats ‚Üí return "Invalid Seat"
    
    If seat is "booked" ‚Üí return "Seat Already Booked"
    
    Otherwise book the seat and return updated seats dictionary

### Example seats dictionary:
{
  "1A": "free",
  "1B": "booked",
  "2A": "free"
}

```python
def book_seat(seats, seat_requested):
    if seat_requested not in seats:
        return "Invalid Seat"
    elif seats[seat_requested] == "booked":
        return "Already Booked"
    else:
        seats[seat_requested] = "booked"
        return seats
seats = {"1A": "free", "1B": "booked", "2A": "free"}
print(book_seat(seats, "1A"))
```

## An online store calculates final prices dynamically:

### Function calculate_discount(price, category) ‚Üí category-based discounts:

    Electronics > $1000 ‚Üí 15%
    
    Clothing > $500 ‚Üí 10%
    
    Groceries ‚Üí 5%
    
    Function apply_coupon(price, coupon_code) ‚Üí nested function applies coupon:
    
    SAVE10 ‚Üí $10 off
    
    SAVE20 ‚Üí $20 off if price > $200
    
    Function final_price(price, category, coupon) ‚Üí combines discount + coupon + tax (8%)

### üì• Input Example:
    final_price(1200, "Electronics", "SAVE10")

### üìù Example Output:
    Price after discount: 1020
    Price after coupon: 1010
    Final Price with Tax: 1090.8

```python
def calculate_discount(price, category):
    if category == "Electronics" and price > 1000:
        return price * 0.85  
    elif category == "Clothing" and price > 500:
        return price * 0.90  
    elif category == "Groceries":
        return price * 0.95 
    return price
def apply_coupon(price, coupon_code):
    def coupon_discount():
        if coupon_code == "SAVE10":
            return 10
        elif coupon_code == "SAVE20" and price > 200:
            return 20
        return 0
    return price - coupon_discount()
def final_price(price, category, coupon):
    discounted_price = calculate_discount(price, category)
    coupon_price = apply_coupon(discounted_price, coupon)
    final = coupon_price * 1.08
    print("Price after discount:", discounted_price)
    print("Price after coupon:", coupon_price)
    print("Final Price with Tax:", final)
final_price(1200, "Electronics", "SAVE10")
```

    Function additional_services(services) ‚Üí Breakfast=$20, Spa=$50, Gym=$10

    total_bill(4, "Deluxe", ["Breakfast", "Spa"], 6)

### üì• Input Example:
    final_price(500, "Electronics", "SAVE10")

### üìù Example Output:
    Price after base discount: 475
    Price after category discount: 427.5
    Price after coupon: 417.5
    Final Price: 451.3 (with 8% tax)

```python
def final_price(price, category, coupon):
    def base_discount(p): 
        return p * 0.95 
    def category_discount(p, c): 
        rates = {"Electronics": 0.10, "Clothing": 0.08, "Grocery": 0.03}
        return p * (1 - rates.get(c, 0))
    def coupon_discount(p, code): 
        return p - 10 if code == "SAVE10" else p
    p1 = base_discount(price)
    p2 = category_discount(p1, category)
    p3 = coupon_discount(p2, coupon)
    final = p3 * 1.08 
    return p1, p2, p3, final
p1, p2, p3, final = final_price(500, "Electronics", "SAVE10")
print("Price after base discount:", p1)
print("Price after category discount:", p2)
print("Price after coupon:", p3)
print("Final Price:", round(final, 1))
```
